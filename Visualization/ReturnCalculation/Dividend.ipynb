# BUY and HOLD
with calucualte dividend

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime, timedelta
```

```python
ticker = 'NVDA'
start_date = '2024-06-01'
end_date = '2025-06-10'
interval = 'monthly' # 'daily', 'weekly', 'monthly','quarterly','yearly'
```

```python
df=yf.download(tickers=ticker,start=start_date, end=end_date, multi_level_index=False, progress=False)
```

```python
df.tail()
```

```python
initial_investment = 300
buy_price = df['Close'].iloc[0]
amount_of_stock = initial_investment/buy_price
dividen_per_one_stock = 0.009
final_price = df['Close'].iloc[-1]

start_date = df.index[0]
end_date = df.index[-1]
holding_period = (end_date - start_date).days

print(f"Initial investment: {initial_investment}")
print(f"Buy stock at {buy_price:.2f}")
print(f"Amount of stock bought: {amount_of_stock:.2f}")
print(f"Holding period: {holding_period} days")
print(f"Divident received per time: {dividen_per_one_stock*amount_of_stock}")

if interval == 'daily':
    n_stock = amount_of_stock
    dividend = dividen_per_one_stock * n_stock * holding_period
    total_return = dividend + final_price*amount_of_stock
    print(f"Divident times received: {holding_period}")
    print(f"Total Dividend: {dividend:.2f}")
    print(f"Final price: {final_price:.2f}")
    print(f"Final return: {total_return} with profit {(total_return-initial_investment)}")
elif interval == 'weekly':
    n_stock = amount_of_stock
    dividend = dividen_per_one_stock * n_stock * (holding_period//7)
    total_return = dividend + final_price*amount_of_stock
    print(f"Divident times received: {holding_period//7}")
    print(f"Total Dividend: {dividend:.2f}")
    print(f"Final price: {final_price:.2f}")
    print(f"Final return: {total_return} with profit {(total_return-initial_investment)}")
elif interval == 'monthly':
    n_stock = amount_of_stock
    dividend = dividen_per_one_stock * n_stock * (holding_period//30)
    total_return = dividend + final_price*amount_of_stock
    print(f"Divident times received: {holding_period//30}")
    print(f"Total Dividend: {dividend:.2f}")
    print(f"Final price: {final_price:.2f}")
    print(f"Final return: {total_return} with profit {(total_return-initial_investment)}")
elif interval == 'quarterly':
    n_stock = amount_of_stock
    dividend = dividen_per_one_stock * n_stock * (holding_period//90)
    total_return = dividend + final_price*amount_of_stock
    print(f"Divident times received: {holding_period//90}")
    print(f"Total Dividend: {dividend:.2f}")
    print(f"Final price: {final_price:.2f}")
    print(f"Final return: {total_return} with profit {(total_return-initial_investment)}")
elif interval == 'yearly':
    n_stock = amount_of_stock
    dividend = dividen_per_one_stock * n_stock * (holding_period//365)
    total_return = dividend + final_price*amount_of_stock
    print(f"Divident times received: {holding_period//365}")
    print(f"Total Dividend: {dividend:.2f}")
    print(f"Final price: {final_price:.2f}")
    print(f"Final return: {total_return} with profit {(total_return-initial_investment)}")
else:
    print("Invalid interval")
```
