# SMA Analysis

```python
#default
ticker="AMZN"
start_date="2024-01-01"
end_date="2025-05-20"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import ta
```

```python
print(f"Ticker: {ticker}")
print(f"Start date: {start_date}")
print(f"End date: {end_date}")
```

```python
# Download historical data
data = yf.download(ticker, start=start_date, end=end_date,multi_level_index=False)
data.reset_index(inplace=True)
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
```

```python
display(data.head())
display(data.tail())
```

```python
#plot the stock price
plt.figure(figsize=(14, 7))
plt.plot(data['Close'])
plt.title(f'{ticker} Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid()
plt.show()
```

## Simple Moving Average (SMA)
- Simple Moving Average (SMA)      SMA 

###  SMA:

 SMA  (Period)   SMA 5   SMA 20  

 SMA :

- SMA = () / ()

  SMA 5  :

1.  5 :  5  10, 12, 11, 13,  14 
2. : 10 + 12 + 11 + 13 + 14 = 60
3. : 60 / 5 = 12
 SMA 5  12 
###  SMA :

- : SMA   SMA    SMA  
- : SMA   SMA   SMA 
- :  SMA   SMA 

```python
#Simple Moving Average (SMA)
data['SMA_5'] = data['Close'].rolling(window=5).mean()
data['SMA_20'] = data['Close'].rolling(window=20).mean()
data['SMA_50'] = data['Close'].rolling(window=50).mean()
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Close Price')
plt.plot(data['SMA_5'], label='5-Day SMA', color='orange')
plt.plot(data['SMA_20'], label='20-Day SMA', color='green')
plt.plot(data['SMA_50'], label='50-Day SMA', color='red')
plt.title(f'{ticker} Stock Price with SMA')
plt.xlabel('Date')
plt.ylabel('Price') 
plt.legend()
plt.grid()
plt.show()
```
