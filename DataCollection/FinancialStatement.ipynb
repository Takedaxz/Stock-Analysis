```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
ticker_symbol = "MSFT"
stock = yf.Ticker(ticker_symbol)
```

```python
income_statement_annual = stock.financials
income_statement_quarterly = stock.quarterly_financials
income_statement = pd.concat([income_statement_annual, income_statement_quarterly], axis=1)
income_statement = income_statement.T
income_statement = income_statement.sort_index()
income_statement.index = income_statement.index.date
income_statement = income_statement.loc[~income_statement.index.duplicated(keep='first')]
income_statement = income_statement.replace(np.NaN, None)
unimportant_columns = [
    'Tax Effect Of Unusual Items',
    'Tax Rate For Calcs',
    'Reconciled Depreciation',
    'Reconciled Cost Of Revenue',
    'Diluted NI Availto Com Stockholders',
    'Basic Average Shares',
    'Operating Revenue',
    'Other Gand A',
    'Net Income Including Noncontrolling Interests',
    'Net Income From Continuing Operation Net Minority Interest',
    'Net Income Continuous Operations',
    'Net Income From Continuing And Discontinued Operation',
    'Interest Expense Non Operating',
    'Interest Income Non Operating',
    'Special Income Charges',
    'Write Off',
    'Gain On Sale Of Security',
    'Total Unusual Items',
    'Total Unusual Items Excluding Goodwill'
]
columns_to_drop = [col for col in unimportant_columns if col in income_statement.columns]
income_statement_cleaned = income_statement.drop(columns=columns_to_drop)
income_statement_cleaned
```

```python
cash_flow_annual = stock.cashflow
cash_flow_quarterly = stock.quarterly_cashflow
cash_flow = pd.concat([cash_flow_annual, cash_flow_quarterly], axis=1)
print(f"Combined Cash Flow Statement for {ticker_symbol}:\n")
cash_flow = cash_flow.T
cash_flow = cash_flow.sort_index()
cash_flow.index = cash_flow.index.date
cash_flow = cash_flow.loc[~cash_flow.index.duplicated(keep='first')]
cash_flow = cash_flow.replace(np.NaN, None)
cash_flow
```

```python
balance_sheet_annual = stock.balance_sheet
balance_sheet_quarterly = stock.quarterly_balance_sheet
balance_sheet = pd.concat([balance_sheet_annual, balance_sheet_quarterly], axis=1)
print(f"Combined Balance Sheet for {ticker_symbol}:\n")
balance_sheet = balance_sheet.T
balance_sheet = balance_sheet.sort_index()
balance_sheet.index = balance_sheet.index.date
balance_sheet = balance_sheet.loc[~balance_sheet.index.duplicated(keep='first')]
balance_sheet = balance_sheet.replace(np.NaN, None)
balance_sheet
```
